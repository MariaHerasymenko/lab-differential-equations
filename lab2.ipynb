## Знайти під яким кутом  $\alpha$ фазові траекторії перетинають пряму $y=x$
$$Q(x, y)=\dot{y}$$$$P(x, y)=\dot{x}
                                    $$$$\frac{d y}{d x}=A=\frac{Q}{P}$$

$$\tan(\alpha) = A \Rightarrow \alpha = \arctan A$$

```python
from sympy import atan ,pi
from sympy.abc import x, y 

P = lambda y,x : 2*y - x 
Q = lambda x,y : 3*x - y
A = Q(x,x)/P(x,x)
alpha = float((atan(A)  * 360) / (2 * pi))
print(f'alpha = {alpha:.4} degrees')
```

1.2 Наша точка (0,0) є ізольованою спеціальною точкою, тому у нас немає пересічних ліній на цій ділянці

1.3 Зобразити фазові траекторії


```python
import numpy as np
import matplotlib.pyplot as plt

def f(Y, t):
    y1, y2 = Y
    return [2*y2-y1, 3*y1-y2]

y1 = np.linspace(-10.0, 10.0, 30)
y2 = np.linspace(-5.0, 5.0, 30)

Y1, Y2 = np.meshgrid(y1, y2)

t = 0

u, v = np.zeros(Y1.shape), np.zeros(Y2.shape)

NI, NJ = Y1.shape

for i in range(NI):
    for j in range(NJ):
        x = Y1[i, j]
        y = Y2[i, j]
        yprime = f([x, y], t)
        u[i,j] = yprime[0]
        v[i,j] = yprime[1]
     

Q = plt.quiver(Y1, Y2, u, v, color='r')

plt.xlabel('$y_1$')
plt.ylabel('$y_2$')
plt.xlim([-10, 10])
plt.ylim([-5, 5])
```

$$
\begin{array}{c}
(x, y)=(0,-1): \\
\lambda_{1,2}= 1  \Longrightarrow \text {Вироджений вузол } \\
(x, y)=(2,-3): \\
\lambda_{1,2}= 1 \pm  \frac {2 \sqrt{3}}{3}   \Longrightarrow \text {Сідло} 
\end{array}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

dx = lambda x,y : x + y + 1
dy = lambda x,y : y + sqrt(2*x**2 + 1)
fig = plt.gcf()
fig.set_size_inches(20, 10)
xticks = np.arange(-7, 8, 1)
yticks = np.arange(-10, 10, 1)
plt.axis((-7, 8, -10, 10))
ax = plt.gca()
ax.set_xticks(xticks)
ax.set_yticks(yticks)

m = np.arange(-6, 7)
n = np.arange(-6, 7)
X, Y = np.meshgrid(m, n)
U = list()
V = list()
for i in m :
    for j in n :
        U.append(dx(i,j))
        V.append(dy(i,j))

ax.quiver(X, Y, U, V, units='xy', width=0.05, scale=60)

plt.show()
```
